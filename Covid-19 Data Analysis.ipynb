```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Load dataset
data = pd.read_csv(r"C:\Users\micha\Downloads\archive (2)\owid-covid-data.csv")
data.head()
```

```python
# Filter for South Africa
country = "South Africa"
df = data[data['location'] == country]

# Select useful columns
df = df[['date', 'total_cases', 'new_cases', 'total_deaths']]

# Convert date column
df['date'] = pd.to_datetime(df['date'])

# Clean data (remove missing values)
df = df.dropna()

# Reset index
df = df.reset_index(drop=True)

# Display first rows
print("Cleaned Data:")
df.head()

```

```python
 #Trend 1: Total Cases Over Time
plt.figure()
plt.plot(df['date'], df['total_cases'])
plt.title("Total Covid-19 Cases in South Africa")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# ðŸ“Š Trend 2: Daily New Cases
# -------------------------------
plt.figure()
plt.plot(df['date'], df['new_cases'])
plt.title("Daily New Covid-19 Cases in South Africa")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# -------------------------------
# ðŸ“‰ Trend 3: Total Deaths
# -------------------------------
plt.figure()
plt.plot(df['date'], df['total_deaths'])
plt.title("Total Deaths in South Africa")
plt.xlabel("Date")
plt.ylabel("Total Deaths")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# -------------------------------
# ðŸ’‰ Correlation: Vaccination vs Cases
# -------------------------------
correlation = df['new_cases'].corr(df['total_deaths'])
print("\nCorrelation between Vaccination and New Cases:", correlation)

# Scatter plot
plt.figure()
plt.scatter(df['new_cases'], df['total_deaths'])
plt.title("Deaths vs New Cases")
plt.xlabel("New cases")
plt.ylabel("Total deaths")
plt.tight_layout()
plt.show()
```
